Lane Choice: Content Decay / Page Refresh Prediction

Why: Content marketing teams and SEO agencies have fixed budgets and limited writer bandwidth. They cannot rewrite every page on a website every month. By identifying which high-traffic pages are genuinely at risk of decaying, we can tell them exactly where to spend their hours to protect their search traffic before it drops off a cliff.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

* The Search Question: Which currently stable SEO pages are most likely to experience a significant traffic decline in the next 90 days if left untouched?
* Unit of Analysis: A single web page URL.
* The Output: A probability score (0.0 to 1.0) indicating the risk of imminent traffic decay.
* The Decision: Which pages should be pulled into the upcoming content rewrite sprint?
* The Action: An SEO strategist assigns the top 50 highest-risk pages to a copywriter to manually update the content, refresh links, and add new information.
* Cost of a False Positive (Predicting decay when it wouldn't happen): The agency wastes money paying a writer to update a page that was already perfectly stable.
* Cost of a False Negative (Missing a page that actually decays): The page loses its rankings on Google, organic traffic vanishes, and the client loses revenue.
* Why data/ML helps: A simple static rule like "refresh any page older than 180 days" is too rigid. It causes too many false positives. Machine learning can balance how traffic volume, ranking position stability, and content age interact dynamically to make a much smarter guess.


## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

raw_github_url = "https://raw.githubusercontent.com/airbender69/FlyRank_ML_Internship/main/data/raw/content_refresh_anonymized.csv"

try:
    df = pd.read_csv(raw_github_url)
    print("Success! File loaded perfectly.\n")

    # Number 1: Total volume of pages tracked
    print(f"Total Pages Tracked: {len(df)}")

    # Number 2: Class breakdown (How many are actually decaying vs growing)
    if 'trend_direction' in df.columns:
        print("\nTraffic Trend Distribution:")
        print(df['trend_direction'].value_counts())

    # Number 3: The Word Count Paradox stats
    if 'word_count' in df.columns and 'trend_direction' in df.columns:
        print("\nAverage Word Count by Traffic Trend:")
        print(df.groupby('trend_direction')['word_count'].mean().round(1))

except Exception as e:
    print(f"Failed to stream the file. Error: {e}")
```

### Data Summary & Core Numbers

An analysis of the starter dataset yields two definitive numbers that prove content decay is a critical operational threat that cannot be solved by simple human assumptions:

1. **High Decay Prevalence:** Out of 30,000 tracked pages, a staggering **16,262 pages (~54.2%)** are actively in a downward traffic trend (`down`). Content decay is not an isolated edge case; it represents the majority state of the current portfolio.
2. **The Word Count Paradox:** Traditional SEO rules of thumb suggest that long-form articles are safer from traffic drops. However, the data reveals that declining pages have a *higher* average word count (**3,221.8 words**) than growing pages (**2,998.1 words**).

Because a simple rule like "longer content is safer" is fundamentally contradicted by the data, a human strategist cannot accurately guess which pages will decay. This non-linear relationship makes building an ML predictive model highly valuable and well worth the next 7 weeks of development.

## 4. Careful words: what I can and can't claim

## Section 4: Careful Words: What I Can and Can't Claim

### What This Work CAN Claim (Observed, Directional, Decision-Support)
* **Observed Statistical Patterns:** The model identifies real, historical correlations within our dataset—such as the relationship between high word counts, positioning stability, and subsequent traffic drops.
* **Directional Risk Scores:** The output provides a directional probability (a risk score between 0.0 and 1.0) indicating whether a page's current performance trajectory mirrors past pages that experienced decay.
* **Decision-Support Tool:** This system functions strictly as an operational prioritization tool. It does not replace human judgment; instead, it helps an SEO strategist decide where to allocate finite writing budgets efficiently.

### What This Work CANNOT Claim (Causal Proof, 'Predicting Google')
* **No Causal Proof:** The model cannot prove *why* a page is losing traffic. It maps correlations, not causation. We cannot claim a specific feature (like a specific word count or CTR drop) directly caused Google to demote a page.
* **Not 'Predicting Google's Algorithm':** This model does not reverse-engineer or predict Google’s proprietary ranking algorithms. It only analyzes internal performance data.
* **No Guarantee of Recovery:** We cannot claim that executing a content refresh based on the model's flags will guarantee a 100% recovery or a return to position one on search engine results pages (SERPs).
